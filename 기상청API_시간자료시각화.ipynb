```python
# Python3   #

import requests
import pprint
import json

url = 'http://apis.data.go.kr/1360000/AsosHourlyInfoService/getWthrDataList'
para ={'serviceKey' : 's9d24DID8McX9iHiTGEso0ztWsa4PWO7VTdsHYHrWLcLZWsVU8SWuuPHlsbg8/b0Zc80XqjGWuiP1Y+xKsxCzw==', 'pageNo' : '1', 'numOfRows' : '500', 'dataType' : 'JSON', 'dataCd' : 'ASOS', 'dateCd' : 'HR', 'startDt' : '20230101', 'startHh' : '01', 'endDt' : '20230501', 'endHh' : '01', 'stnIds' : '99' }

res = requests.get(url, params=para)
print(res.content)
```

JSON LOAD

```python
import json
json_file = json.loads(res.text)
json_file
```

```python
import pandas as pd
from pandas import json_normalize
df = json_normalize(json_file['response']['body']['items']['item'])
```

```python
import matplotlib.pyplot as plt

df['tm']=pd.to_datetime(df['tm'])
df['ts']=pd.to_numeric(df['ts'], downcast='float')

fig = plt.figure(figsize=(14,6))
ax = plt.subplot(1,1,1)
ax.plot(df['tm'],df['ts'],color='r',lw=1.5,alpha=0.7)

ax.set_ylabel('Temperature ', fontweight='bold', fontsize=12)
ax.set_xlabel('Date ', fontweight='bold', fontsize=12)
ax.set_title('OPEN API ASOS  data', fontweight = 'bold',fontsize = 14)
```

```python
import matplotlib.pyplot as plt

df['tm']=pd.to_datetime(df['tm'])
df['ta']=pd.to_numeric(df['ta'], downcast='float')

fig = plt.figure(figsize=(14,6))
ax = plt.subplot(1,1,1)
ax.plot(df['tm'],df['ta'],color='b',lw=1.5,alpha=0.7)

ax.set_ylabel('AIRTemperature ', fontweight='bold', fontsize=12)
ax.set_xlabel('Date ', fontweight='bold', fontsize=12)
ax.set_title('OPEN API ASOS AIR  ', fontweight = 'bold',fontsize = 14)
```

```python
# 
fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1,1,1)
ax.plot (df['tm'],df['ts'], color='blue', lw=1.5, alpha=0.7, label='')
ax.plot (df['tm'],df['ta'], color='red', lw=1.5, alpha=0.7, label='AIR ')
ax.legend()
plt.show()
```
